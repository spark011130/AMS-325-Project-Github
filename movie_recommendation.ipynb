```python
from sklearn.feature_extraction.text import CountVectorizer

# Converting list objects to strings: space separated (tasks to vectorize text data)
data_df['genres_literal'] = data_df['genres'].apply(lambda x : (' ').join(x))
# ex. ['Action', 'Adventure'] -> 'Action Adventure'
# 0 = Action
# 1 = Adventure

# CountVectorizer (Use cosine similarity to compare genre similarity by taking data into the form of a vector)
count_vect = CountVectorizer(min_df=1, ngram_range=(1,2))  # ngram = word combination
genre_mat = count_vect.fit_transform(data_df['genres_literal']).toarray()

print(genre_mat.shape)
```

```python
def cosine_similarity(matrix):
    """
    Computes cosine similarity manually between rows of a matrix.
    """
    dot_product = np.dot(matrix, matrix.T)  # Compute dot product for every pair of rows
    norm = np.linalg.norm(matrix, axis=1)  # Compute the norm (magnitude) of each row
    similarity = dot_product / (norm[:, None] * norm[None, :])  # Compute cosine similarity
    np.fill_diagonal(similarity, 1.0)  # Set diagonal to 1 (self-similarity)
    return similarity

# Compute cosine similarity manually
genre_sim = cosine_similarity(genre_mat)
# Output for five movies (first index 5)
genre_sim[:5]
```

```python
def movie_recommendation(df, sim_matrix, title_name, top_n=10):

    # Find the index of the movie
    title_movie = df[df['title'] == title_name]
    title_index = title_movie.index.values

    # Add a film's similarity data frame (sim_matrix: similarity matrix)
    df["similarity"] = sim_matrix[title_index, :].reshape(-1,1)

    # Extract the index after sorting the similarity in descending order
    temp = df.sort_values(by="similarity", ascending=False) # Sort in order of highest similarity
    final_index = temp.index.values[ : top_n]

    return df.iloc[final_index]
```

```python
# 10 movies with high genre similarities to "Avatar"
similar_movies = movie_recommendation(data_df, genre_sim, 'Avatar', 10)
similar_movies[['title', 'rating_average', 'rating_count', "similarity"]]
```

```python
# To avoid these 3 cases of sampling bias, we added weighted evaluation functions.
```

```python
# weighted_rating evaluation function
# v: the total number of movie ratings, R: the average rating of the movie, m: Minimum number of rating counts for the upper percentile, C: the average rating of an entire movie
percentile = 0.7  # the minimum number of rating count for a movie

# 'rating_count' Converting columns to numeric (error resolution)
data_df['rating_count'] = pd.to_numeric(data_df['rating_count'], errors='coerce')

# 'rating_average' converting columns to numeric
data_df['rating_average'] = pd.to_numeric(data_df['rating_average'], errors='coerce')

m = data_df['rating_count'].quantile(percentile)  # films with top 30 percent of the rating average
C = data_df['rating_average'].mean()  # the average of rating_average

def weighted_rating_average(record):
    v = record['rating_count']
    R = record['rating_average']

    return ( (v/(v+m)) * R ) + ( (m/(m+v)) * C )

data_df['rat_count'] = data_df.apply(weighted_rating_average, axis=1)
```

```python
# weighted_popularity evaluation function
percentile = 0.7  # The minimum number of rating count for a movie

m = data_df['popularity'].quantile(percentile)  # Top 30 percent of popular movies
C = data_df['rating_average'].mean()  # The average of rating_average

def weighted_popularity(record):
    v = record['popularity']
    R = record['rating_average']

    return ( (v/(v+m)) * R ) + ( (m/(m+v)) * C )

data_df['rat_popularity'] = data_df.apply(weighted_popularity, axis=1)
```

```python
# weighted_revenue evaluation function
percentile = 0.7  # The minimum number of rating count for a movie

m = data_df['revenue'].quantile(percentile)  # movies with top 30 percent of revenue
C = data_df['rating_average'].mean()

def weighted_revenue(record):
    v = record['revenue']
    R = record['rating_average']

    return ( (v/(v+m)) * R ) + ( (m/(m+v)) * C )

data_df['rat_revenue'] = data_df.apply(weighted_revenue, axis=1)
```

```python
class MovieRecommender:
    def __init__(self, df, sim_matrix):
        self.df = df.copy()
        self.sim_matrix = sim_matrix

    def find_sim_movie_with_rating_number(self, title_name, top_n=10, rat_weight=0.6, sim_weight=0.4):
        """
        Top movies are recommended based on similarity and ratings (with rating_count top 30%).

        Parameters:
        - title_name: movie title
        - top_n: the number of recommended movies
        - rat_weight: weighted average rating (default value: 0.6)
        - sim_weight: weighted similarity  (default value: 0.4)
        """
        title_movie = self.df[self.df['title'] == title_name]
        title_index = title_movie.index.values[0]

        # Add similarity to the dataframe
        self.df["similarity"] = self.sim_matrix[title_index, :].reshape(-1)

        # calculate the maximum value of 'rat_count' for normalization and calculate weighted_rat_count
        max_rat_count = self.df["rat_count"].max()
        self.df["weighted_rat_count"] = (
            sim_weight * self.df["similarity"] + rat_weight * (self.df["rat_count"] / max_rat_count)
        )

        # Except for the movies chosen by the customer
        df_filtered = self.df[self.df.index != title_index]

        # Select the top n after sorting in descending order based on weighted_rat_count
        recommended_movies = df_filtered.sort_values(by="weighted_rat_count", ascending=False).head(top_n)

        return recommended_movies

    def find_sim_movie_with_popularity(self, title_name, top_n=10, rat_weight=0.6, sim_weight=0.4):
        """
        Top movies are recommended based on similarity and ratings (with popularity top 30%).
        """
        title_movie = self.df[self.df['title'] == title_name]
        title_index = title_movie.index.values[0]

        self.df["similarity"] = self.sim_matrix[title_index, :].reshape(-1)

        max_rat_popularity = self.df["rat_popularity"].max()
        self.df["weighted_rat_popularity"] = (
            sim_weight * self.df["similarity"] + rat_weight * (self.df["rat_popularity"] / max_rat_popularity)
        )

        df_filtered = self.df[self.df.index != title_index]
        recommended_movies = df_filtered.sort_values(by="weighted_rat_popularity", ascending=False).head(top_n)

        return recommended_movies

    def find_sim_movie_with_revenue(self, title_name, top_n=10, rat_weight=0.6, sim_weight=0.4):
        """
        Top movies are recommended based on similarity and ratings (with revenue top 30%).
        """
        title_movie = self.df[self.df['title'] == title_name]
        title_index = title_movie.index.values[0]

        self.df["similarity"] = self.sim_matrix[title_index, :].reshape(-1)

        max_revenue = self.df["rat_revenue"].max()
        self.df["weighted_rat_revenue"] = (
            sim_weight * self.df["similarity"] + rat_weight * (self.df["rat_revenue"] / max_revenue)
        )

        df_filtered = self.df[self.df.index != title_index]
        recommended_movies = df_filtered.sort_values(by="weighted_rat_revenue", ascending=False).head(top_n)

        return recommended_movies
```

```python
# Example: 'Avatar'
# A table in which the top 30% rating counts are listed in the order of highest ratings
# The proportion of genre similarities and ratings is 4 to 6
recommender = MovieRecommender(data_df, genre_sim)
recommended = recommender.find_sim_movie_with_rating_number('Avatar', top_n=10, rat_weight=0.6, sim_weight=0.4)
recommended[["title", "rating_average", "rating_count", "rat_count", "similarity"]]
```
